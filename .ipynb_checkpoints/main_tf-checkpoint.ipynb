# Import libraries

```python
# Reload modules
%load_ext autoreload
%autoreload 2
# Inline plots
%matplotlib inline

# Standard
import numpy as np
import matplotlib.pyplot as plt
import time as time
from copy import deepcopy

# tf
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

# Mine
from utility import display_image, permutate
```

# Set parameters

```python
# Training
N_task = 3       # Number of tasks
N_epoch = 1      # Number of epochs
batch_size = 100 # Number of samples in each minibatch

# Network
hidden_size = 50       # Number of hidden layer neurons

# Regularization and Fisher
lambda_L2 = 0.2        # Regularization parameter for L2
lambda_EWC = 3         # Regularization parameter for EWC
sample_size_Fish = 200 # Number of samples to use to estimate Fisher Info.

# Miscellaneous
```

# Load data

```python
# Extract data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)

# Non-permuted datasets
# X = images, y = labels (one-hot vectors)
X_train_list = [mnist.train.images]    # 55000 x 784
y_train_list = [mnist.train.labels]    # 55000 x 10
X_val_list = [mnist.validation.images] # 5000 x 784
y_val_list = [mnist.validation.labels] # 5000 x 10
X_test_list = [mnist.test.images]      # 10000 x 784
y_test_list = [mnist.test.labels]      # 10000 x 10
```

# Create permutated data

```python
X_list = [X_train_list[0],X_val_list[0],X_test_list[0]]

for i in range(N_task):
    [X_tr,X_v,X_te],_ = permutate(X_list)
    
    X_train_list.append(X_tr)
    X_val_list.append(X_v)
    X_test_list.append(X_te)
    
    y_train_list.append(deepcopy(y_train_list[0]))
    y_val_list.append(deepcopy(y_val_list[0]))
    y_test_list.append(deepcopy(y_test_list[0]))
```

# Plot sample images

```python
sample_index = 10      # Sample image to display
# Display label
label = y_train_list[i][sample_index]
label = np.nonzero(label)[0][0]
print('label = ' + str(label))

# Display images
plt.figure()
for i in range(N_task):
    image = X_train_list[i][sample_index]

    plt.subplot(1,N_task,1+i)
    display_image(image)
```
